```python
import matplotlib.pyplot as plt
import numpy as np
```

# Basic Line Plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 15, 25, 40])

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# Line Styles and Colors

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle='dashdot', color='green', marker='o')
plt.title("Styled Line Example")
plt.show()
```

# Multiple Lines in One Graph with Numpy and Matplotlib

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])

plt.plot(x, y1, marker='s', color='red', label='Line 1')
plt.plot(x, y2, marker='o', color='green', label='Line 2', linestyle='--')

plt.title("Multiple Line Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.show()

```

# Bar Chart with Numpy and Matlotlib

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([5, 9, 1, 7])

plt.bar(x, y, color='orange')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Scatter Plot with Numpy and Matplotlib

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 6, 3, 7])

plt.scatter(x, y, color='purple', s=100)
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# Histogram with Numpy and Matplotlib

```python
data = np.array([3, 4, 5, 6, 8, 8, 8, 9, 10])

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Count")
plt.show()
```

# Subplots(Multiple Graphs Together)

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))   # 8 => width, 4 => height

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title("Subplot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title("Subplot 2")

plt.tight_layout()
plt.show()
```

# Saving Charts

```python
x = np.array([1, 2, 3, 4])
y = np.array([4, 8, 6, 10])

plt.plot(x, y, color='green')
plt.title("Save Figure Example")
plt.savefig('my_plot.png')
plt.show()
```
